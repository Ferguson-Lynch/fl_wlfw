```python
import pandas as pd
```

```python
all_species_excel = pd.ExcelFile('All Taxa Species Lists 07152019 (1).xlsx')
```

```python
species_status = pd.DataFrame()

for animal_type in all_species_excel.sheet_names:
    print(animal_type)
    sheet = pd.read_excel(all_species_excel, animal_type)
    if 'RSGCN Status' in sheet.columns:
        species_status = pd.concat([species_status, sheet[['Scientific Name', 'RSGCN Status']]])
```

```python
# There is a problem in the data. After researching, the species seems to not be in danger at all
species_status[species_status[['Scientific Name']].duplicated(keep=False)]
```

```python
species_status = species_status.drop(species_status[(species_status['Scientific Name'] == 'Caracara cheriway')
                                   & (species_status['RSGCN Status'] == 'A) Yes')].index)
```

```python
species_status[species_status[['Scientific Name']].duplicated(keep=False)]
```

```python
watershed_data = pd.read_excel('WorkingCopyOfWatershedData_ForRSGCN.xlsx')
```

```python
species_status['Scientific Name']
```

```python
merged = watershed_data.merge(species_status, left_on='gname', right_on='Scientific Name', how='left')
```

```python
merged[['RSGCN Status']] = merged[['RSGCN Status']].fillna('No Match')
```

```python
merged = merged.drop(columns=['Scientific Name'])
```
